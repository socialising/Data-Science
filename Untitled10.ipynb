```python
import pandas as pd

# Load education datasets for classification and regression tasks
df_classification_education = pd.read_csv('education_classification_dataset.csv')
df_regression_education = pd.read_csv('education_regression_dataset.csv')

# Check the first few rows of each dataset
print("Classification Dataset:")
print(df_classification_education.head())
print("\nRegression Dataset:")
print(df_regression_education.head())
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Data Cleaning: Handle missing values, duplicates, and outliers
def clean_data(df):
    df.fillna(df.mean(numeric_only=True), inplace=True)  # Handle missing values by imputing with mean
    df.drop_duplicates(inplace=True)    # Remove duplicates
    return df

df_classification_education = clean_data(df_classification_education)
df_regression_education = clean_data(df_regression_education)

# Data Integration: Merge and preprocess the two selected datasets
# Assuming 'exam_scores' is the target variable for both datasets
X_classification_education = df_classification_education.drop('exam_scores', axis=1)
y_classification_education = df_classification_education['exam_scores']

X_regression_education = df_regression_education.drop('exam_scores', axis=1)
y_regression_education = df_regression_education['exam_scores']

# Feature Selection/Engineering and Data Transformation
numeric_features_classification = X_classification_education.select_dtypes(include=['float64']).columns
categorical_features_classification = X_classification_education.select_dtypes(include=['object']).columns

numeric_features_regression = X_regression_education.select_dtypes(include=['float64']).columns
categorical_features_regression = X_regression_education.select_dtypes(include=['object']).columns

# Preprocessing pipelines
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder())
])

preprocessor_classification = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features_classification),
        ('cat', categorical_transformer, categorical_features_classification)
    ])

preprocessor_regression = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features_regression),
        ('cat', categorical_transformer, categorical_features_regression)
    ])
```

```python
# Step 5: Predictive Analysis - Regression

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Check if regression dataset is empty before proceeding
if not df_regression_education.empty:
    # Separate features and target variable for regression
    X_regression_education = df_regression_education.drop('exam_scores', axis=1)
    y_regression_education = df_regression_education['exam_scores']

    # Split data into training and testing sets
    X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
        X_regression_education, y_regression_education, test_size=0.2, random_state=42
    )

    # Preprocess categorical features using one-hot encoding
    categorical_features = X_regression_education.select_dtypes(include=['object']).columns
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(), categorical_features)
        ],
        remainder='passthrough'
    )

    X_train_reg_preprocessed = preprocessor.fit_transform(X_train_reg)
    X_test_reg_preprocessed = preprocessor.transform(X_test_reg)

    # Initialize regression model
    best_model_reg = RandomForestRegressor()

    # Train the regression model
    best_model_reg.fit(X_train_reg_preprocessed, y_train_reg)

    # Predict on test data
    y_pred_reg = best_model_reg.predict(X_test_reg_preprocessed)

    # Print regression evaluation metrics
    mse = mean_squared_error(y_test_reg, y_pred_reg)
    r2 = r2_score(y_test_reg, y_pred_reg)
    print(f"Regression Model Performance: \nMean Squared Error: {mse:.4f}, R-squared: {r2:.4f}")

else:
    print("Regression dataset is empty after cleaning. Cannot proceed with regression analysis.")
```
