```python
from sklearn.datasets import load_wine, load_iris, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load datasets
wine = load_wine()
iris = load_iris()
cancer = load_breast_cancer()

# Function to train and evaluate a Decision Tree classifier
def train_and_evaluate(dataset, name):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        dataset.data, dataset.target, test_size=0.2, random_state=42
    )

    # Initialize Decision Tree classifier
    dt_classifier = DecisionTreeClassifier(random_state=42)

    # Train the classifier
    dt_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = dt_classifier.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)

    print(f"Accuracy on {name} dataset: {accuracy:.2f}")

    # Plot the Decision Tree
    plt.figure(figsize=(12, 8))
    plot_tree(dt_classifier, filled=True, feature_names=dataset.feature_names, class_names=dataset.target_names)
    plt.title(f'Decision Tree for {name} dataset')
    plt.show()

# Train and evaluate Decision Tree on Wine dataset
train_and_evaluate(wine, "Wine")

# Train and evaluate Decision Tree on Iris dataset
train_and_evaluate(iris, "Iris")

# Train and evaluate Decision Tree on Breast Cancer dataset
train_and_evaluate(cancer, "Breast Cancer")
```
